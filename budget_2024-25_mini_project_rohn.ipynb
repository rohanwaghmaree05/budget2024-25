```python
import openpyxl
import pandas as pd

income = openpyxl.Workbook()
sheet = income.active

data = [
    ['Tax Slab', 'FY 2023-24 Rate', 'FY 2024-25 Rate'],
    ['Upto ₹ 3 lakh', 'Nil', 'Nil'],
    ['₹ 3 lakh - ₹ 6 lakh', '5%', '-'],
    ['₹ 3 lakh - ₹ 7 lakh', '-', '5%'],
    ['₹ 6 lakh - ₹ 9 lakh', '10%', '-'],
    ['₹ 7 lakh - ₹ 10 lakh', '-', '10%'],
    ['₹ 9 lakh - ₹ 12 lakh', '15%', '-'],
    ['₹ 10 lakh - ₹ 12 lakh', '-', '15%'],
    ['₹ 12 lakh - ₹ 15 lakh', '20%', '20%'],
    ['More than ₹ 15 lakh', '30%', '30%']
]   # Define the data for the tax slabs

for row in data:  # Append the data to the sheet
    sheet.append(row)

    income.save('data.xlsx')  # Save the income to a file

    tax = pd.read_excel('data.xlsx')  # Read the data back using pandas
    tax
```

```python

tax_slabs = [
    "Upto ₹ 3 lakh", "₹ 3 lakh - ₹ 6 lakh", "₹ 3 lakh - ₹ 7 lakh",
    "₹ 6 lakh - ₹ 9 lakh", "₹ 7 lakh - ₹ 10 lakh", "₹ 9 lakh - ₹ 12 lakh",
    "₹ 10 lakh - ₹ 12 lakh", "₹ 12 lakh - ₹ 15 lakh", "More than ₹ 15 lakh"
]
rate_2023_24 = [0, 5, 0, 10, 0, 15, 0, 20, 30]
rate_2024_25 = [0, 0, 5, 0, 10, 0, 15, 20, 30]

x = np.arange(len(tax_slabs))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, rate_2023_24, width, label='FY 2023-24')
bars2 = ax.bar(x + width/2, rate_2024_25, width, label='FY 2024-25')

ax.set_xlabel('Tax Slab')
ax.set_ylabel('Rates (%)')
ax.set_title('Comparison of Income Tax Slabs under New Regime Before and After Budget')
ax.set_xticks(x)
ax.set_xticklabels(tax_slabs, rotation=80, ha='right')
ax.legend()

plt.tight_layout()
plt.show()
```

```python
data = [
    ['Country', 'Grants and Loans (in crore Rs)'],
    ['Bhutan', 2398.97],
    ['Maldives', 770.90],
    ['Nepal', 650.00],
    ['Myanmar', 370.00],
    ['Mauritius', 330.00],
    ['Afghanistan', 220.00],
    ['Bangladesh', 130.00],
    ['Sri Lanka', 60.00],
    ['Seychelles', 9.91],
    ['Mongolia', 5.00]
]


workbook = openpyxl.Workbook()
sheet = workbook.active


for row in data:
    sheet.append(row)


workbook.save('grants_and_loans.xlsx')


grants_and_loans = pd.read_excel('grants_and_loans.xlsx')

countries = grants_and_loans['Country'][1:]  # Create the pie chart
amounts = grants_and_loans['Grants and Loans (in crore Rs)'][1:]  

plt.figure(figsize=(10, 8))
plt.pie(amounts, labels=countries, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio is confirmed pie is drawn as a circle.
plt.figtext(0.5, 0.01, 'Grants and Loans Provided by Indian Government (2024-25)', ha='center', fontsize=12)
plt.show()
```
